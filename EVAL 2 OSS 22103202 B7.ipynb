```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pymongo import MongoClient
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client['rehab_facility']
collection = db['provider']

data = {
    "CMS Certification": [],
    "Provider Name": [],  
    "Address Line 1": [], 
    "Address Line 2": [], 
    "City/Town": [], 
    "State": [], 
    "ZIP Code": [], 
    "County/Parish": [], 
    "Telephone Number": [], 
    "CMS Region": [], 
    "Ownership Type": [],
    "Certification Date": []
}

df = pd.DataFrame(data)
df = pd.read_csv('Inpatient_Rehabilitation.csv')

# collection.insert_many(df.to_dict('records'))

for idx, row in df.iterrows():
    data = {
    "CMS Certification": row["CMS Certification Number (CCN)"],
    "Provider Name": row["Provider Name"],
    "Address Line 1": row["Address Line 1"], 
    "Address Line 2": row["Address Line 2"], 
    "City/Town": row["City/Town"],
    "State": row["State"], 
    "ZIP Code": row["ZIP Code"], 
    "County/Parish": row["County/Parish"], 
    "Telephone Number": row["Telephone Number"], 
    "CMS Region": row["CMS Region"], 
    "Ownership Type": row["Ownership Type"],
    "Certification Date": row["Certification Date"],
    }
    collection.insert_one(data)

non_profit = df[(df['Ownership Type'] == 'Non-Profit') & (df['Certification Date'] > '2011-10-01')]

birmingham_profit = df[(df['City/Town'] == 'BIRMINGHAM') & (df['Ownership Type'] == 'For profit')]

zip_range = df[(df['ZIP Code'] >= 85000) & (df['ZIP Code'] <= 90000)]

provider_counts = df['Ownership Type'].value_counts()
profit_count = provider_counts.get('For profit', 0)
non_profit_count = provider_counts.get('Non-profit', 0)
```

```python
non_profit_list = non_profit.to_dict(orient='records')

birmingham_profit_list = birmingham_profit.to_dict(orient='records')

zip_range_list = zip_range.to_dict(orient='records')

print("\nFor-Profit providers are : ", profit_count)
print("\nNon-Profit providers are : ", non_profit_count)

print("\nList all Non-Profit providers that were certified after 10-01-2011 \n")
for p in non_profit_list:
    print(p)
    
print("\nList all providers belonging to BIRMINGHAM and Ownership Type= PROFIT \n")
for p in birmingham_profit_list:
    print(p)
    
print("\nProviders with ZIP codes in the range 85000-90000 \n")
for p in zip_range_list:
    print(p)

```
